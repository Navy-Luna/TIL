## Grouping by One Column

```python
# From https://www.kaggle.com/code/sanjanabasu/tips-dataset
tips = Table.read_table('tips.csv')
tips
```

```python
# 유난히 금요일에만 손님이 안모이는 느낌이 있는걸 분석해낼 수 있음
tips.select('day', 'tip').group('day')
```

```python
# group 뒤에 옵션을 넣어서 count만이 아니라 평균도 해볼 수 있음.
tips.select('day', 'tip').group('day', np.average)
```

```python
# 숫자 이외에는 평균내지는 않음
# 다양한 형태로 해석을 해볼 수 있음
tips.group('day', np.average)

# size를 보면 일요일날 3명에 가까움 -> 그래서 그것에 비례하여 tip과 total_bill이 높아지는 것이라고 유추 가능
```

```python
tips.select('day', 'tip').group('day', np.min)
```

```python
# 끝에 list를 쓰면 group에 포함되는 리스트를 돌려준다.
tips.select('day', 'tip').group('day', list)
```

## Pivot Tables

```python
tips
```

```python
# pivot tabl을 사용하여 행으로 보여줄 애들, 열로 보여줄 애들을 선택하여
# pivot('열이 될 축', '행이 될 축')
tips.pivot('day','sex')
```

```python
# 뒤에 optional로 values와 collect 설정 가능
# 아무것도 없으면 행, 열에 해당하는 count값
# 둘다 넣거나 둘다 빼거나 둘 중 하나임
tips.pivot('day','sex',values='total_bill',collect=np.sum)
```

## Discussion Questions

```python
sky = Table.read_table('skyscrapers.csv')
sky = (sky.with_column('age', 2022 - sky.column('completed'))
            .drop('completed'))
sky.show(3)
```

```python
# 1. For each city, what's the tallest building for each material?
sky_pivot = sky.pivot('material','city')
sky_pivot
```

```python
sky_pivot = sky.pivot('material','city' 
                      ,values='height',collect=np.max)
sky_pivot
```

```python
# 2. For each city, what's the height difference between the tallest
#    steel building and the tallest concrete building?
sky_pivot = sky_pivot.with_column(
    'difference',
    abs(sky_pivot.column('steel')) - abs(sky_pivot.column('concrete'))
)
sky_pivot
```

```python
# 3. Generate a table of the names of the oldest buldings for each
#    material for each city.
def first(s):
    "Return the first element in an array."
    return s.item(0)

(sky
 .sort('age', descending=True)
 .pivot('material', 'city', 'name', first)
)#.where('city', 'San Francisco')
```

## Joins

```python
drinks
```

```python
discounts = Table().with_columns(
    'Coupon % off', make_array(10, 25, 5),
    'Location', make_array('Asha', 'Strada', 'Asha')
)
discounts
```

```python
# join을 사용하여 서로 매칭되는 항목에 값들을 추가해서 합칠 수 있음
# 아래는 location과 cafe에 동일한 항목들을 서로 합친 것임
combined = drinks.join('Cafe', discounts, 'Location')
combined
```

```python
discounted_frac = 1 - combined.column('Coupon % off') / 100
combined.with_column(
    'Discounted Price', 
    combined.column('Price') * discounted_frac
)
```

하나의 카페에서 두 개의 음료를 구입한다면, 동일 table의 동일 'Cafe'로 join하면 새로운 table을 생성할 수 있다.

동일한 column label이 있기 때문에 두번째 것은 "_2"가 레이블에 붙게 된다.
