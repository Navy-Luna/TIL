# 패딩

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
preprocessed_sentences = [['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(preprocessed_sentences)
encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

여기까지 저번에 했던 정수 인코딩 작업

```python
max_len = max(len(item) for item in encoded)
print('최대 길이 :',max_len)
```

```python
# 모든 문장의 길이를 최대 길이의 문장과 같게 맞춤
for sentence in encoded:
    while len(sentence) < max_len:
        sentence.append(0)

padded_np = np.array(encoded)
padded_np
```

## 케라스 도구로 패딩하기

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

```python
padded = pad_sequences(encoded)
padded
```

```python
# pad_sequences는 기본적으로 앞에 0을 채움
padded = pad_sequences(encoded, padding='post')
padded
```

```python
padded = pad_sequences(encoded, padding='post', maxlen=5) # maxlen 설정해서 자르기도 함
padded # 이것도 기본적으로 앞에서 자름
```

```python
padded = pad_sequences(encoded, padding='post', truncating='post', maxlen=5)
padded # truncate 옵션을 사용하여 뒤의 단어를 삭제할 수 있다.
```

```python
last_value = len(tokenizer.word_index) + 1 # 단어 집합의 크기보다 1 큰 숫자를 사용
print(last_value)
```

```python
# 단어집합보다 하나 큰 숫자를 이용하여 padding할때 0이 아닌 다른 숫자로 패딩을 한 모습
padded = pad_sequences(encoded, padding='post', value=last_value)
padded
```

# 원-핫 인코딩

- 단어집합 : 서로 다른 단어들의 집합니다. 기본적으로 book와 books와 같은 단어도 변형 형태로 다른 단어라고 간주한다. 텍스트의 모든 단어들을 중복을 허용하지 않고 모아놓으면 이를 단어 집합이라고 한다. 각 단어에 숫자를 부여하고 이를 벡터로 다루고 싶다면 어떻게 해야할까?

- 원-핫인코딩 : 단어 집합의 크기를 벡터의 차원이라고 하고 표현 하고 싶은 단어의 인덱스에 1의 값을 부여하고 다른 인덱스에는 0을 부여하는 단어의 벡터 표현 방식을 나타낸다.

```python
!pip install konlpy
```

```python
# 한국어의 원-핫 인코딩을 진행해보자
from konlpy.tag import Okt

okt = Okt()
tokens = okt.morphs("나는 자연어 처리를 배운다")
print(tokens)
```

```python
word_to_index = {word : index for index, word in enumerate(tokens)} # 이렇게 enumerate를 사용한다.
print('단어 집합 :',word_to_index)
```

```python
def one_hot_encoding(word, word_to_index):
  one_hot_vector = [0]*(len(word_to_index)) # 인덱스 위치 빼고는 모두 0
  index = word_to_index[word]
  one_hot_vector[index] = 1
  return one_hot_vector
```

```python
one_hot_encoding("자연어", word_to_index)
```

## keras를 이용해서 원-핫 인코딩을 해보자

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical # util 안에 있음

text = "나랑 점심 먹으러 갈래 점심 메뉴는 햄버거 갈래 갈래 햄버거 최고야"

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text]) # 한국어 문장 생으로 넣을때는 이렇게 하는듯
print('단어 집합 :',tokenizer.word_index)
```

```python
sub_text = "점심 먹으러 갈래 메뉴는 햄버거 최고야"
encoded = tokenizer.texts_to_sequences([sub_text])[0]
print(encoded) # 다른 걸로 인코딩
```

```python
one_hot = to_categorical(encoded)
print(one_hot) # 각 숫자가 원-핫 인코딩된 형태로 나뉜다.
```

원-핫 인코딩은 단어간의 유사성 관계를 반영하기가 힘들고 용량을 많이 먹는다.

위의 원-핫 인코딩의 문제점을 해결하기위해 단어의 잠재적 의미를 반영하여 다차원 공간에 벡터화 하는 방법이 2가지 있다.

1. 카운트 기반의 벡터화 방법 (LSA, HAL)
2. 예측 기반으로 하는 방법 (NNLM, RNNLM, Word2Vec, FastText)
3. 둘 다 사용하는 GloVe
