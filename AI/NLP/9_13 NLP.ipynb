# 불용어 (Stopword)

유의미한 단어 토큰만을 골라내기 위해서는 큰 의미가 없는 단어 토큰을 제거하는 작업이 필요하다. 예를 들면, I, my, me, 접미사 같은 단어들은 문장에서는 자주 등장하지만 분석하는데 큰 도움이 되지는 않음. 이런 단어를 우리는 "불용어"라고 한다. NLTK에서는 100개 이상의 영단어를 불용어로 패키지내에 미리 정의하고 있음

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from konlpy.tag import Okt
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words_list = stopwords.words('english') # NLTK가 정의한 영어 불용어 리스트를 리턴
print('불용어 개수 :', len(stop_words_list))
print('불용어 10개 출력 :',stop_words_list[:10])
```

```python
# NLTK를 사용하여 불용어 삭제하기
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(example)

result = []
for word in word_tokens:
  if word not in stop_words:
    result.append(word)

print('불용어 제거 전 :',word_tokens)
print('불용어 제거 후 :',result)
```

```python
# 한국어를 사용하여 불용어를 정의하고 제거해보자
# not in 사용하여 제거

okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨대 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전 :',word_tokens)
print('불용어 제거 후 :',result)
```

# 정수 인코딩

자연어 처리에서는 텍스트를 숫자로 바꾸는 여러가지 기법이 존재함. 그러한 기법을 바꾸는 첫 방법으로 각 단어를 고유한 정수에 매핑(mapping)시키는 전처리 작업이 필요한 때가 있음

보통은 단어 등장 빈도를 기준으로 정렬한 뒤, 숫자를 할당함

```python
# Dictionary 사용하기
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```python
# 문장 토큰화
sentences = sent_tokenize(raw_text)
print(sentences)
```

```python
# 정제 작업과 정규화 작업을 병행하여 단어 토큰화를 수행한다.
# 단어들을 소문자화하여 단어의 개수 통일
# 불용어와 단어 길이 2보다 작으면 없애기
# 단어가 텍스트일때 최대한의 전처리 작업을 끝내둬야함

vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
  # 단어 토큰화
  tokenized_sentence = word_tokenize(sentence)
  result = []

  for word in tokenized_sentence:
    word = word.lower() # 모든 단어를 소문자화하여 단어의 개수를 줄인다.
    if word not in stop_words:
      if len(word) > 2:
        result.append(word)
        if word not in vocab: # dict에 없으면
          vocab[word] = 0 # 단어 추가
        vocab[word] += 1

  preprocessed_sentences.append(result)
print(preprocessed_sentences)
```

```python
print("단어 집합 :", vocab)
```

```python
# 정렬 -> value 기준 정렬 (알아두기)
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
print(vocab_sorted)
```

```python
word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted:
  if frequency > 1: # 빈도수가 작은 단어는 제외
    i = i + 1
    word_to_index[word] = i

print(word_to_index)
```

자연어 처리를 하다보면 텍스트 데이터에 있는 모든 단어를 사용한다기보단 빈도수가 높은 단어들을 많이 사용함. 보통 상위 n개 단어를 사용... 여기서는 상위 5개단어만 사용해보자

```python
vocab_size = 5

# 인덱스가 5초과인 단어 제거
words_frequency = [word for word, index in word_to_index.items() if index >= vocab_size + 1] # 한줄 표현식 good

# 일단 위에서 value가 5보다 큰 단어들 골라옴
# 해당 단어에 대한 인덱스 정보를 삭제
for w in words_frequency:
  del word_to_index[w]

print(word_to_index)
```

```python
# 이제 문장에 있던 단어들을 각각의 숫자로 바꿀 것임
# 예를 들어 ['barber', 'person']은 [1, 5]가 된다. 하지만 ['barber', 'good', 'person']은 없는 단어 'good'가 있음
# 없는 애들은 Out of vocabulary라고하고 새롭게 인덱스를 추가해주자
word_to_index['OOV'] = len(word_to_index) + 1
print(word_to_index) # 없는 단어를 뜻하는 OOV를 따로 추가했음
```

```python
encoded_sentences = []
for sentence in preprocessed_sentences:
    encoded_sentence = []
    for word in sentence:
        try:
            # 단어 집합에 있는 단어라면 해당 단어의 정수를 리턴.
            encoded_sentence.append(word_to_index[word])
        except KeyError: # 이런 방식으로 keyerror를 이용!
            # 만약 단어 집합에 없는 단어라면 'OOV'의 정수를 리턴.
            encoded_sentence.append(word_to_index['OOV'])
    encoded_sentences.append(encoded_sentence)
print(encoded_sentences)
```

## 텍스트 처리할때 케라스 사용하기 (제일 편한듯)

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# 전처리는 알아서 해야함
preprocessed_sentences = [['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
```

```python
tokenizer = Tokenizer()

# fit_on_texts()안에 코퍼스를 입력으로 하면 빈도수를 기준으로 단어 집합을 생성.
# fit_on_texts()는 입력한 텍스트로부터 단어 빈도수가 높은 순으로 낮은 정수 인덱스를 부여한다
tokenizer.fit_on_texts(preprocessed_sentences)
```

```python
print(tokenizer.word_index) # 이렇게 하면 상위 몇개 쓰기가 힘든듯..? 짜를 순 있음 word_index로 확인 가능하긴함
```

```python
print(tokenizer.texts_to_sequences(preprocessed_sentences)) # 알아서 인코딩
```

```python
# 위에 Counter나 FreqDist처럼 상위 몇개 단어만 사용할 수  있음
# Tokenizer(num_words=숫자)로 사용 가능
vocab_size = 5
tokenizer = Tokenizer(num_words = vocab_size + 1) # 상위 5개 단어만 사용.. 1을 더 더해줘야함
tokenizer.fit_on_texts(preprocessed_sentences) # 0부터 산정함

# 자연어 처리의 패딩작업때문에 0이 존재함.. 케라스 토크나이즈를 사용할 때는 숫자 0도 단어 집합의 크기로 고려해야함
```
